```python
'''
USE THIS CODE!
Convert images using MATLAB V7.3 formatting.
'''

import os
import numpy as np
import cv2
import h5py

def convert_images_to_mat(input_folder, output_folder):
    # Create output folders if they don't exist
    os.makedirs(output_folder, exist_ok=True)
    
    # List all JPEG image files in the input folder
    image_files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.jpg', '.jpeg'))]
    
    # Process each image file
    for filename in image_files:
        # Set up image path
        image_path = os.path.join(input_folder, filename)
        
        # Read images; make sure they are grayscale
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  
        
        # Check if image was successfully read
        if image is not None:
            # Create data structure for .mat file
            data_dict = {
                'cjdata': {
                    'PID': np.array([], dtype=np.uint16),  # Example PID (change as needed)
                    'image': image.astype(np.int16),  # Convert image to int16
                    'label': np.array([3], dtype=np.float64),  # Example label (change as needed)
                    'tumorBorder': np.zeros((1, 38), dtype=np.float64),  # Placeholder for tumorBorder
                    'tumorMask': np.zeros_like(image, dtype=np.uint8)  # Placeholder for tumorMask
                }
            }
            
            # Create path for the output .mat file
            mat_filename = os.path.splitext(filename)[0] + '.mat'
            mat_file_path = os.path.join(output_folder, mat_filename)
            
            # Save data dictionary into a .mat file using h5py
            with h5py.File(mat_file_path, 'w') as file:
                for key, value in data_dict.items():
                    group = file.create_group(key)
                    for subkey, subvalue in value.items():
                        group.create_dataset(subkey, data=subvalue)
            
            # Print confirmation message
            print(f"Converted {image_path} to {mat_file_path}")
        else:
            # Print error message if image reading failed
            print(f"Failed to read image: {image_path}")

# Specify the input and output folders
input_folder = 'C:/Users/embilab/OneDrive - Emory University/Spring 2024/CS - 534 Machine Learning/Machine Learning Project/SARTAJ_Dataset/pituitary_tumor'
output_folder = 'C:/Users/embilab/OneDrive - Emory University/Spring 2024/CS - 534 Machine Learning/Machine Learning Project/Benchmark/All Together'

# Convert images to MATLAB v7.3 .mat files
convert_images_to_mat(input_folder, output_folder)

```

```python
#  ALTER FOR MATLAB V7.3
import os
import gzip
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
import h5py

def convert_nii_gz_to_jpeg_and_mat(input_folder, output_folder):
    # Create output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Initialize index counter
    file_count = sum(1 for file in os.listdir(input_folder) if file.endswith('.nii.gz'))
    index = 1  # Start index from 1

    # Iterate over all directories and subdirectories in the input folder
    for root, dirs, files in os.walk(input_folder):
        for filename in files:
            if filename.endswith('t1.nii.gz'):  # Check for .nii.gz files ending with 't1.nii.gz'
                try:
                    # Construct full path to the current file
                    input_filepath = os.path.join(root, filename)

                    # Unzip the .nii.gz file to a temporary .nii file
                    temp_nii_filepath = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}.nii")
                    with gzip.open(input_filepath, 'rb') as f_in:
                        with open(temp_nii_filepath, 'wb') as f_out:
                            f_out.write(f_in.read())

                    # Load the NIfTI (.nii) file
                    nii_img = nib.load(temp_nii_filepath)
                    nii_data = np.array(nii_img.get_fdata(), dtype=np.int16)

                    # Check image dimensions and extract relevant data
                    if len(nii_data.shape) == 3:
                        image = nii_data[:, :, nii_data.shape[2] // 2]  # Extract middle slice for 3D image
                    elif len(nii_data.shape) == 4:
                        image = nii_data[:, :, :, :3]  # Extract RGB channels for 4D image
                    else:
                        raise ValueError(f"Unsupported image dimensions: {nii_data.shape}")

                    # Generate filenames based on index
                    jpeg_filename = f"g({index}).jpg"
                    mat_filename = f"g({index}).mat"
                    jpeg_filepath = os.path.join(output_folder, jpeg_filename)
                    mat_filepath = os.path.join(output_folder, mat_filename)

                    # Convert image to JPEG and save
                    plt.imsave(jpeg_filepath, image, cmap='gray')

                    # Prepare data dictionary for saving into .mat file
                    data_dict = {
                        'PID': np.array([1, 2, 3, 4, 5, 6], dtype=np.uint16), #not necessary
                        'image': image.astype(np.int16),
                        'label': np.array([1.0], dtype=np.float64),
                        'tumorBorder': np.zeros((1, 38), dtype=np.float64),
                        'tumorMask': (image > 0).astype(np.uint8)
                    }

                    # Save data to .mat file in v7.3 format
                    with h5py.File(mat_filepath, 'w') as mat_file:
                        for key, value in data_dict.items():
                            mat_file.create_dataset(f'cjdata/{key}', data=value)

                    print(f"Converted {filename} to {jpeg_filename} and saved to {output_folder}")
                    print(f"Saved corresponding .mat file to {mat_filepath}")

                    # Increment index for the next file
                    index += 1

                except Exception as e:
                    print(f"Error processing {filename}: {e}")

                finally:
                    # remove temp
                    if os.path.exists(temp_nii_filepath):
                        os.remove(temp_nii_filepath)

    print("Conversion of .nii.gz files to JPEG and .mat format completed.")

# Define input and output directories
input_folder = 'C:/Users/embilab/OneDrive - Emory University/Spring 2024/CS - 534 Machine Learning/Machine Learning Project/BRaTS18_Dataset/LGG'
output_folder = 'C:/Users/embilab/OneDrive - Emory University/Spring 2024/CS - 534 Machine Learning/Machine Learning Project/Benchmark/Glioma'


# Function call to convert 'T1.nii.gz' files in each directory of the input folder
convert_nii_gz_to_jpeg_and_mat(input_folder, output_folder)

```

# CHECK THAT BOTH DATA SETS HAVE THE SAME VARIABLE TYPES

```python
import os
import h5py

def inspect_hdf5_object(obj, name=''):
    """ Recursively inspect HDF5 object (dataset or group). """
    if isinstance(obj, h5py.Dataset):
        # If object is a dataset, print dataset information
        print(f" - Variable '{name}': Type - {obj.dtype}, Shape - {obj.shape}")
    elif isinstance(obj, h5py.Group):
        # If object is a group, recursively inspect its contents
        print(f" - Group '{name}'")
        for key in obj.keys():
            inspect_hdf5_object(obj[key], name + '/' + key)

def inspect_mat_file(mat_file_path):
    try:
        # Open MATLAB v7.3 file using h5py
        with h5py.File(mat_file_path, 'r') as file:
            # Print information about variables in the HDF5 file
            print(f"Variables in file: {mat_file_path}")
            for key in file.keys():
                inspect_hdf5_object(file[key], key)
    except Exception as e:
        print(f"Error inspecting {mat_file_path}: {e}")

# Specify the directory containing MATLAB v7.3 .mat files
mat_files_dir = 'C:/Users/embilab/OneDrive - Emory University/Spring 2024/CS - 534 Machine Learning/SVM/data_even'

# List all MATLAB v7.3 .mat files in the directory
mat_files = [f for f in os.listdir(mat_files_dir) if f.endswith('.mat')]

# Inspect each MATLAB v7.3 .mat file using h5py
for mat_file in mat_files:
    mat_file_path = os.path.join(mat_files_dir, mat_file)
    inspect_mat_file(mat_file_path)

```
